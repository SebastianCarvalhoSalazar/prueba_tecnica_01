### Manejo de Outliers
Se ha observado la presencia de datos atípicos en los diferentes features del dataset , por lo tanto, aquellos puntos fuera del límite inferior y superior del boxplot correspondiente se remplazarán por np.NaN para posteriormente imputar en ellos un estadístico, esto debido a que eliminar las filas que los contengan no es conveniente para el dataset ya que se reduciría drásticamente su tamaño.

```python
headers_filter = headers
from scipy.stats import iqr

for header in range(len(headers_filter)):
    Q1 = df_filtered[headers_filter[header]].quantile(0.25) # Percentil 25
    Q3 = df_filtered[headers_filter[header]].quantile(0.75) # Percentil 75
    IQR = iqr(df_filtered[headers_filter[header]], axis = 0) # Rango Intercuartil
    lim_sup = Q3 + 1.5 * IQR # Limite Superior
    lim_inf = Q1 - 1.5 * IQR # Limite Inferior
    df_filtered[headers_filter[header]][(df_filtered[headers_filter[header]]>lim_sup)|(df_filtered[headers_filter[header]]<lim_inf)] = np.nan
```

### Imputación de Datos

```python
headers = df_filtered.columns[1:]

from sklearn.impute import SimpleImputer
imputer_1 = SimpleImputer(missing_values=np.nan, strategy='median')
imputer_0 = SimpleImputer(missing_values=np.nan, strategy='median')

for header in range(len(headers)):
    data_1 = np.array(df_filtered[headers[header]][df_filtered['diagnosis'] == 1]).reshape(-1,1)
    data_0 = np.array(df_filtered[headers[header]][df_filtered['diagnosis'] == 0]).reshape(-1,1)
    
    stat_1 = imputer_1.fit(data_1).statistics_
    stat_0 = imputer_0.fit(data_0).statistics_

    df_filtered[headers[header]][(np.isnan(df_filtered[headers[header]]) & (df_filtered['diagnosis'] == 1))] = stat_1
    df_filtered[headers[header]][(np.isnan(df_filtered[headers[header]]) & (df_filtered['diagnosis'] == 0))] = stat_0
```
