```python
mapping = {'M': 1, 'B': 0}
df_filtered["diagnosis"] = df_filtered["diagnosis"].map(mapping)
```

```python
headers
```

```python
headers_filter = headers
from scipy.stats import iqr

for header in range(len(headers_filter)):
    Q1 = df_filtered[headers_filter[header]].quantile(0.25)
    Q3 = df_filtered[headers_filter[header]].quantile(0.75)
    IQR = iqr(df_filtered[headers_filter[header]], axis = 0)
    lim_sup = Q3 + 1.5 * IQR
    lim_inf = Q1 - 1.5 * IQR
    df_filtered[headers_filter[header]][(df_filtered[headers_filter[header]]>lim_sup)|(df_filtered[headers_filter[header]]<lim_inf)] = np.nan
```

```python
headers = df_filtered.columns[1:]

from sklearn.impute import SimpleImputer
imputer_1 = SimpleImputer(missing_values=np.nan, strategy='median')
imputer_0 = SimpleImputer(missing_values=np.nan, strategy='median')

for header in range(len(headers)):
    data_1 = np.array(df_filtered[headers[header]][df_filtered['diagnosis'] == 1]).reshape(-1,1)
    data_0 = np.array(df_filtered[headers[header]][df_filtered['diagnosis'] == 0]).reshape(-1,1)
    
    stat_1 = imputer_1.fit(data_1).statistics_
    stat_0 = imputer_0.fit(data_0).statistics_

    df_filtered[headers[header]][(np.isnan(df_filtered[headers[header]]) & (df_filtered['diagnosis'] == 1))] = stat_1
    df_filtered[headers[header]][(np.isnan(df_filtered[headers[header]]) & (df_filtered['diagnosis'] == 0))] = stat_0
```

```python
# generamos nuevas columnas con ruido aleatorio
rng = np.random.RandomState(42)
ruido = rng.normal(size=(df_filtered.shape[0], len(df_filtered.columns)))
df_ruido = pd.DataFrame(ruido)
df_ruido.head()
```

```python
df_filtered_2 = pd.concat([df_filtered,df_ruido],axis=1)
df_filtered_2.head()
```

```python
y = df_filtered_2['diagnosis']                # Variable dependiente (a predecir)
X = df_filtered_2.drop(['diagnosis'], axis=1) # Variables independientes (features)
```

```python
from sklearn.feature_selection import f_classif        
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import SelectPercentile
```

```python
select = SelectPercentile(score_func=f_classif, percentile=50)
```

```python
select.fit(X, y)
```

```python
X_sel = select.transform(X)
print("X_train",X.shape) 
print("X_train_sel",X_sel.shape)
```

```python
mask = select.get_support()
```

```python
plt.figure(figsize=(20,6))
plt.matshow(mask.reshape(1, -1), cmap='gray_r')
plt.xlabel("indice")
plt.show()
```

```python
Feature_Selection = [a for a in zip(X.columns,mask[:len(df_filtered.columns)-1])]
Feature_Selection
```

```python
fil = [Feature_Selection[i][0] for i in range(len(df_filtered.columns)-1) if Feature_Selection[i][1] == True]
fil
```

```python
df_filtered_reduced = df_filtered[fil]
```
