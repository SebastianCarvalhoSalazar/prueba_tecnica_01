### Manejo de Outliers
Se ha observado la presencia de datos atípicos en los diferentes features del dataset , por lo tanto, aquellos puntos fuera del límite inferior y superior del boxplot correspondiente se remplazarán por np.NaN para posteriormente imputar en ellos un estadístico, esto debido a que eliminar las filas que los contengan no es conveniente para el dataset ya que se reduciría drásticamente su tamaño.

```python
headers_filter = headers
from scipy.stats import iqr

for header in range(len(headers_filter)):
    Q1 = df_filtered[headers_filter[header]].quantile(0.25) # Percentil 25
    Q3 = df_filtered[headers_filter[header]].quantile(0.75) # Percentil 75
    IQR = iqr(df_filtered[headers_filter[header]], axis = 0) # Rango Intercuartil
    lim_sup = Q3 + 1.5 * IQR # Limite Superior
    lim_inf = Q1 - 1.5 * IQR # Limite Inferior
    df_filtered[headers_filter[header]][(df_filtered[headers_filter[header]]>lim_sup)|(df_filtered[headers_filter[header]]<lim_inf)] = np.nan
```

### Mediana Estadística por Feature y Categoría

El estadístico seleccionado para la imputación de los datos faltantes es la mediana, esto debido a que es más robusta en comparación al promedio, ya que esta se ve menos afectada por la presencia de sesgos en la distribución de los datos o de valores extremos, sin embargo, ya que el dataset cuenta con dos clasificaciones (M,B), se calculó una mediana para cada feature dependiendo de su clasificación.

### Imputación de Datos

```python
headers = df_filtered.columns[1:]

from sklearn.impute import SimpleImputer
imputer_1 = SimpleImputer(missing_values=np.nan, strategy='median')
imputer_0 = SimpleImputer(missing_values=np.nan, strategy='median')

for header in range(len(headers)):
    data_1 = np.array(df_filtered[headers[header]][df_filtered['diagnosis'] == 1]).reshape(-1,1)
    data_0 = np.array(df_filtered[headers[header]][df_filtered['diagnosis'] == 0]).reshape(-1,1)
    
    stat_1 = imputer_1.fit(data_1).statistics_
    stat_0 = imputer_0.fit(data_0).statistics_

    df_filtered[headers[header]][(np.isnan(df_filtered[headers[header]]) & (df_filtered['diagnosis'] == 1))] = stat_1
    df_filtered[headers[header]][(np.isnan(df_filtered[headers[header]]) & (df_filtered['diagnosis'] == 0))] = stat_0
```

```python
df_filtered = pd.DataFrame(df_filtered, columns=headers_filter) # Se convierten los Datos nuevamente a un DataFrame
df_filtered.sample(5)
```

```python
from scipy.stats import skewtest, levene, shapiro

# probar la hipótesis nula de que la asimetría de la población de la 
# que se extrajo la muestra es la misma que la de una distribución normal 
# correspondiente.
skw = skewtest(df_filtered)
plt.figure(figsize=(14,4))
plt.plot(skw.pvalue)
plt.xticks(range(len(df_filtered.columns)),df_filtered.columns, rotation=90)
plt.ylabel("p-value")
plt.xlabel("Feature")
plt.title("Skewtest")

# probar la hipótesis nula de que todas las muestras de entrada provienen de 
# poblaciones con varianzas iguales

print(levene(df_filtered["area_mean"],df_filtered["perimeter_mean"])) 
print(levene(df_filtered["smoothness_mean"],df_filtered["smoothness_worst"])) 
```

### ANOVA para Reducción Dimensional

```python
# generamos nuevas columnas con ruido aleatorio
rng = np.random.RandomState(42)
ruido = rng.normal(size=(df_filtered.shape[0], len(df_filtered.columns)))
df_ruido = pd.DataFrame(ruido)
df_ruido.head()
```

```python
# Concatenamos las Nuevas Columnas al Dataset
df_filtered_2 = pd.concat([df_filtered,df_ruido],axis=1)
df_filtered_2.head()
```

```python
y = df_filtered_2['diagnosis']                # Variable dependiente (Target)
X = df_filtered_2.drop(['diagnosis'], axis=1) # Variables independientes (features)
```

```python
from sklearn.feature_selection import f_classif         # Calcule el valor F de ANOVA para la muestra      
from sklearn.feature_selection import SelectPercentile 
```

```python
select = SelectPercentile(score_func=f_classif, percentile=50)
```

```python
select.fit(X, y)
```

```python
X_sel = select.transform(X)
print("X_train",X.shape) 
print("X_train_sel",X_sel.shape)
```

```python
mask = select.get_support()
```

### Features Seleccionadas

```python
plt.figure(figsize=(20,6))
plt.matshow(mask.reshape(1, -1), cmap='gray_r')
plt.xlabel("indice")
plt.show()
```

```python
# Features Seleccionadas y No Seleccionadas
Feature_Selection = [a for a in zip(X.columns,mask[:len(df_filtered.columns)-1])]
Feature_Selection
```

```python
# Filtro de Features Seleccionadas
fil = [Feature_Selection[i][0] for i in range(len(df_filtered.columns)-1) if Feature_Selection[i][1] == True]
fil
```
